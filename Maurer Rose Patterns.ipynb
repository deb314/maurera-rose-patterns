```python
from math import *
import matplotlib.pyplot as plt
import numpy as np

x1=[]
y1=[]
n=6

for theta in range(0,361):
    r=cos(radians(n*theta))
    x=r*cos(radians(theta))
    y=r*sin(radians(theta))
    x1.append(x)
    y1.append(y)
    
d=97

x2=[]
y2=[]

for i in range(0,361):
    k=i*d
    r=sin(radians(n*k))
    x3=r*cos(radians(k))
    y3=r*sin(radians(k))
    x2.append(x3)
    y2.append(y3)
    
page=plt.figure(figsize=(7,7))
ax=page.add_axes([0,0,1,1])

ax.plot(x2,y2,'black')
#ax.plot(x1,y1,'r',linewidth=3)
plt.show()

```
